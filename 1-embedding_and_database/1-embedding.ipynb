```python
from langchain_huggingface import HuggingFaceEmbeddings

## initialize simple embedding model no api key needed
embedding=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
    
)
embedding
```

```python
# create your first embedding
text=" hello, i am learning about embedding"

embedding_text=embedding.embed_query(text)
print(f"text:{text}")
print(f"embedding length: {len(embedding_text)}")
print(embedding_text)
```
